### Divide Hero/Publisher column

```python
# Exploring existing format with a few examples
superhero['Hero|Publisher'].head(2)
```

```python
## examining a single value from the coordinates col
info = superhero.loc[0,"Hero|Publisher"]
print(type(info))
info
```

```python
## adding expand=True
superhero['Hero|Publisher'].str.split('|',expand=True)
```

```python
## save the 2 new columns into the dataframe
superhero[['Hero','Publisher']] = superhero['Hero|Publisher'].str.split('|',expand=True)
superhero.head(2)
```

```python
# drop the Hero|Publisher 
superhero = superhero.drop(columns=['Hero|Publisher'])
superhero.head(2)
```

### Divide the measurements column 

```python
# Exploring existing format with a few examples
superhero['Measurements'].head(2)
```

```python
## examining a single value from the measurement col
measure = superhero.loc[0,"Measurements"]
print(type(measure))
measure
```

```python
measure = measure.replace("'",'"')
measure
```

```python
## now we can use json.loads
fixed_measure = json.loads(measure)
print(type(fixed_measure))
fixed_measure
```

```python
## use .str.replace to replace all single quotes
superhero['Measurements'] = superhero['Measurements'].str.replace("'",'"')
## Apply the json.loads to the full column
superhero['Measurements'] = superhero['Measurements'].apply(json.loads)
superhero['Measurements'].head()
```

```python
#create a column from the dictionary
height_weight = superhero['Measurements'].apply(pd.Series)
height_weight
```

```python
# concat height_weight with original dataframe
superhero = pd.concat((superhero, height_weight), axis = 1)
superhero.head(2)
```

```python
#fix the weight column
# Make a list of all characters to replace
to_replace = [' kg',]
# run a loop to replace all of the characters in the list at once
for char in to_replace:
    superhero['Weight'] = superhero['Weight'].str.replace(char,'',regex=False)
    
superhero['Weight'].head(5)
```

```python
# convert to a float
superhero['Weight'] = superhero['Weight'].astype(float)
```

```python
#fix the height column
# Make a list of all characters to replace
to_replace = [' cm',]
# run a loop to replace all of the characters in the list at once
for char in to_replace:
    superhero['Height'] = superhero['Height'].str.replace(char,'',regex=False)
    
superhero['Height'].head(5)
```

```python
# convert to a float
superhero['Height'] = superhero['Height'].astype(float)
```

```python
# drop the original column it was already dropped
superhero = superhero.drop(columns=['Measurements'])
```

```python
# Create a new column where the single quotes are replaced by double quotes
superpower['powers_split'] = superpower['Powers'].str.split(',')
```

```python
## exploding the column of lists
exploded = superpower.explode('powers_split')
exploded.head(20)
```

```python
## saving the unique values from the exploded column
cols_to_make = exploded['powers_split'].dropna().unique()
cols_to_make
```

```python
superpower.head(20)
```

```python
#create new columns
for col in cols_to_make:
    superpower[col] = superpower['Powers'].str.contains(col)
superpower.head()
```

```python
# drop powers columns
superpower = superpower.drop(columns=['Powers','powers_split'])
## save data for next step
superpower.to_csv('superpower_all.csv', index=False)
```

### Combine the dataframes

# Answer the questions

- Compare the average weight of super powers who have Super Speed to those who do not.
- What is the average height of heroes for each publisher?

```python
# check column names
super_merged['Super Speed'].value_counts()
```

```python
#  use groupby to  find average weight of superspeed versus those who do not
super_merged.groupby(['Super Speed'])['Weight'].mean()
```

- The average Weight of non-superspeed is 101.77 kg. 
- The average weight of superspeed is 129.40 kg. 

```python
# using groupby
super_merged.groupby(['Publisher'])['Height'].mean()
```
